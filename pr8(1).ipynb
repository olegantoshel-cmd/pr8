```python
import re
INN_PATTERN = r'^\d{12}$'

def validate_inn():
    while True:
        inn = input("Введите ваш ИНН (12 цифр): ").strip()

        if re.fullmatch(INN_PATTERN, inn):
            print(f" ИНН принят: {inn}")
            return inn
        else:
            print(" Неверный формат ИНН. ИНН должен состоять ровно из 12 цифр.")


user_inn = validate_inn()
print(f"Ввод завершен. ИНН: {user_inn}")

print("\n" + "="*50 + "\n")
```

```python
while True:
    password = input("Введите пароль (8-20 символов, заглавная буква и цифра): ")

    if len(password) < 8 or len(password) > 20:
        print("❌ Ошибка! Длина пароля должна быть от 8 до 20 символов.")
        continue

    has_upper = False
    for char in password:
        if char.isupper():
            has_upper = True
            break

    if not has_upper:
        print("❌ Ошибка! Пароль должен содержать хотя бы одну ЗАГЛАВНУЮ букву.")
        continue

    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break

    if not has_digit:
        print("❌ Ошибка! Пароль должен содержать хотя бы одну цифру.")
        continue

    print("✅ Пароль надежный!")
    password_result = password
    break
```
